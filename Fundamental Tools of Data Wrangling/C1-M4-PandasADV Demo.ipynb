## Aggregation

```python
byArtist = df_subset.groupby('Artist')
```

```python
byArtist
```

```python
byArtist.mean()
```

```python
byArtist.sum()
```

```python
byArtist.sum().sort_values(by = ['Views'], ascending=False)
```

```python
byArtist.describe()
```

```python
type(byArtist.describe())
```
