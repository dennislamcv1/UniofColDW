```python
import pandas as pd
import atoti as tt
```

## Create Session

```python
session = tt.Session()
```

```python
df = pd.read_csv('/content/Spotify_Youtube_Sample.csv')
df.head()
```

```python
views = session.read_csv(
    '/content/Spotify_Youtube_Sample.csv',
    keys = ['Artist', 'Track', 'Album'],
)
views.head()
```

## Create a Data Cube

```python
cube = session.create_cube(views, name = 'Views')
cube
```

```python
m= cube.measures
m
```

```python
l = cube.levels
l
```

## Play with the Cube

### Apex or 0-D

```python
cube.query(m['Views.SUM'])
```

### 1D

```python
cube.query(m['Views.SUM'], levels = [l['Artist']])
```

```python
cube.query(m['Views.SUM'], levels = [l['Album']])
```

### 2D

```python
cube.query(m['Views.SUM'], levels = [l['Album'], l['Artist']])
```

### 2D with Slicing

```python
cube.query(m['Views.SUM'], levels = [l['Album'], l['Track']], filter=l['Artist'] == 'The Beatles')
```

```python
cube.query(m['Views.SUM'], levels = [l['Album'], l['Track']], filter=l['Artist'] == 'Michael Jackson')
```

```python
cube.query(m['Views.SUM'], levels = [l['Album'], l['Track']], filter=l['official_video'] == 'True')
```

```python
cube.query(m['Views.SUM'], levels = [l['Album'], l['Track']], filter=l['official_video'] == 'False')
```

### 3D

```python
cube.query(m['Views.SUM'], levels = [l['Album'], l['Track'], l['official_video']])
```

### 3D with Slicing

```python
cube.query(m['Views.SUM'], levels = [l['Album'], l['Track'], l['official_video']], filter=l['Artist'] == 'Michael Jackson')
```
