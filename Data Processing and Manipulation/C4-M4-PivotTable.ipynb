## Setup

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Spotify_Youtube_Sample.csv')
```

```python
df.info()
```

```python
df['Artist'].unique()[:100]
```

## Create Pivot Table

```python
df_pivot = df.pivot_table(values=['Views','Likes','Comments'],index=['Artist','Album'],aggfunc='mean')
df_pivot
```

## Play with the Pivot Table

```python
df_pivot.loc['Michael Jackson']
```

```python
df_pivot.loc[['Michael Jackson', 'The Beatles', 'Beyoncé']]
```

## Sort the Values

```python
df_pivot.sort_values(by = 'Views', ascending= False)
```

```python
df_pivot.sort_values(by = ['Artist','Views'], ascending= False)
```

```python
df_pivot.sort_values(by = ['Artist','Views'], ascending= False).loc['Daddy Yankee']
```

```python
df_pivot.sort_values(by = ['Artist','Views'], ascending= False).loc[['Coldplay', 'The Beatles']]
```
