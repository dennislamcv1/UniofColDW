## MinMaxScalar Scaling

It is the process of performing a linear transformation on original data. 

$v_i^{'} = \frac{v_i - min}{max - min} (max^{'} - min^{'}) + min^{'}$

where $v_i$ is current value, $min$ and $max$ are the current min and max, and $max^{'}$ and $min^{'}$ are the new boundary. $v_i^{'}$ is the min_max scaled value.

Normally, we use a special case of [0, 1] as the new scale, in this case, the formular can be as simple as:

$v_i^{'} = \frac{v_i - min}{max - min}$

```python
df['population_MinMax'] = (df['population'] - df['population'].min())/ (df['population'].max() - df['population'].min())
df['population_MinMax']
```

```python
df['population_MinMax'].describe()
```

## Z-score Normalization/standardization

In this technique, the values are normalized based on the mean and standard deviation of attribute A. Each value is substracted with the mean, thus, we leave with the variance in terms of standard deviation.

$v_i^{'} = \frac{v_i - mean}{std}$

where $v_i$ is the current v alue, $mean$ and $std$ are current mean and standard deviation, and $v_i^{'}$ is the Z-score scaled value.

```python
df['population_Z'] = (df['population'] - df['population'].mean())/(df['population'].std())
df['population_Z'] 
```

```python
df['population_Z'] .describe()
```

## Decimal Scaling

It normalizes by moving the decimal point of values of the data.

$v_i^{'} = \frac{v_i}{10^{t}}$

where $v_i$ is current value, $t$ is the number of digits of the max absolute value + 1, and $v_i^{'}$ is the decimal scaled value.

Lets understand it by an example: Suppose we have data set in which the value ranges from -1234 to 999.  In this case the maximum absolute value is 1234 with 4 digits, and t is 4 + 1 = 5. 

So to perform decimal scaling, we divide each of values in data set by $10^5$

```python
df['population'].max()
```

```python
df['population_decimal'] = df['population']/100000
df['population_decimal']
```

```python
df['population_decimal'].describe()
```
