## Setup

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Economy_of_US.csv')
df
```

## A simple splot

```python
plt.plot(df['Year'], df['GDP_Growth'])
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

### Change markers

```python
plt.plot(df['Year'], df['GDP_Growth'], 'o')
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

```python
markers = ['o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h', 'v', '^', '<', '>', '1', '2', '3', '4', '|', '_']

for m in markers:
  print(m)
  plt.plot(df['Year'], df['GDP_Growth'], m)
  plt.xlabel('Year')
  plt.ylabel('GDP GRowth')
  plt.title('Economy of US')
  plt.show()
```

### Change line style

```python
line = ['-',':','--','-.']
for l in line:
  print(l)
  l = 'o' + l + 'r'
  plt.plot(df['Year'], df['GDP_Growth'], l)
  plt.xlabel('Year')
  plt.ylabel('GDP GRowth')
  plt.title('Economy of US')
  plt.show()
```

### Change color

```python
color = ['r','g','b','c','m','y','k','w']
for c in color:
  print(c)
  c = 'o:' + c
  plt.plot(df['Year'], df['GDP_Growth'], c)
  plt.xlabel('Year')
  plt.ylabel('GDP GRowth')
  plt.title('Economy of US')
  plt.show()
```

### Change marker size

```python
plt.plot(df['Year'], df['GDP_Growth'], marker = '^', ms = 15)
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

## Scatter plot

```python
plt.plot(df['Year'], df['GDP_Growth'], 'o')
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

```python
plt.scatter(df['Year'], df['GDP_Growth'])
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

### colorbar

```python
plt.scatter(df['Year'], df['GDP_Growth'], c=df['Inflation'], cmap='hot')
plt.colorbar()
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

### size

```python
plt.scatter(df['Year'], df['GDP_Growth'], s= df['Unemployment']*1000)
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

### colorbar and size

```python
plt.scatter(df['Year'], df['GDP_Growth'], c=df['Inflation'], cmap='hot', alpha = 0.5, s= df['Unemployment']*1000)
plt.colorbar()
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

```python
plt.bar(df['Year'], df['GDP_Growth'])
plt.xlabel('Year')
plt.ylabel('GDP GRowth')
plt.title('Economy of US')
```

```python
plt.hist(df['GDP_Growth'])
plt.xlabel('GDP Growth')
plt.ylabel('Counts')
plt.title('Economy of US')
```

```python
plt.hist(df['GDP_Growth'], bins = 50)
plt.xlabel('GDP Growth')
plt.ylabel('Counts')
plt.title('Economy of US')
```

## Pie plot

```python
df['Inflation_Change'].value_counts()
```

```python
plt.pie(df['Inflation_Change'].value_counts(), labels = ['Decrease', 'Increase', 'No change'])
plt.legend()
plt.title('Economy of US')
```

```python
plt.pie(df['Inflation_Change'].value_counts(), labels = ['Decrease', 'Increase', 'No change'], explode = [0.0, 0.2, 0])
plt.legend()
plt.title('Economy of US')
```

## Box plot

```python
plt.boxplot(df['Inflation'])
plt.ylabel('Inflation')
plt.title('Economy of US')
```

## Violinplot

```python
plt.violinplot(df['Inflation'])
plt.ylabel('Inflation')
plt.title('Economy of US')
```

## Multiplots

```python
plt.plot(df['Year'], df['GDP_Growth'])
plt.plot(df['Year'], df['Inflation'])
plt.plot(df['Year'], df['Unemployment'])
plt.xlabel('Year')
plt.title('Economy of US')
```

### Add legend

```python
plt.plot(df['Year'], df['GDP_Growth'], label = 'GDP_Growth')
plt.plot(df['Year'], df['Inflation'], label = 'Inflation' )
plt.plot(df['Year'], df['Unemployment'], label = 'Unemployment')
plt.xlabel('Year')
plt.ylabel('Economy')
plt.grid()
plt.legend()
```

### Change arrangement

```python
plt.subplot(3, 1, 1)
plt.plot(df['Year'], df['GDP_Growth'], label = 'GDP_Growth')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Economy')
plt.subplot(3, 1, 2)
plt.plot(df['Year'], df['Inflation'], label = 'Inflation' )
plt.legend()
plt.xlabel('Year')
plt.ylabel('Economy')
plt.subplot(3, 1, 3)
plt.plot(df['Year'], df['Unemployment'], label = 'Unemployment')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Economy')
plt.grid()
```
